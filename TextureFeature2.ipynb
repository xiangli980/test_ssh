```python
case_names = ["tumor_572304","tumor_572302","tumor_572298","tumor_572297"]

for case_name in case_names:
  df = pd.DataFrame(columns=cols)
  names = os.listdir("./headandneck/{}/results_CD320/mask".format(case_name))

  for name in names:
      #if name == "524266_Interface_1.png":
        #continue
      patch_name = name.split(".")[0]
      imagePath = "./headandneck/{}/results_CD320/topology/{}".format(case_name, name)
      maskPath = "./Inflammtion_model/Dataset/results/mask.png"
        
      result = extractor.execute(imagePath, maskPath, label=1)
      
      lists = result.items()
      add = {}
      add["case"] = case_name
      add["patch"] = patch_name
      for i, (x,y) in enumerate(lists):
        if i>=22:
          add[x] = y
      df =  df.append(add, ignore_index=True)

  df.to_csv("./headandneck/{}/results_CD320/texture_feature.csv".format(case_name), index=False)
```
